```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile
```

```python
with zipfile.ZipFile('archive (2).zip', 'r') as zip_ref:
    zip_ref.extractall('unzipped_data')
```

```python
df = pd.read_csv('/content/unzipped_data/Airbnb_Open_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna(subset=['price', 'room type', 'neighbourhood group'])
df.fillna(df.median(numeric_only=True),inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='room type', data=df, palette='viridis')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['price'])
plt.title('Price Outliers')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(x=df['price'], bins=50, kde=True)
plt.title('Price Distributiion')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Map')
plt.show()
```

```python
df['price'] = df['price'].str.replace('[\$,]', '', regex=True).astype(float)
```

```python
summary={
    "Total Listings": len(df),
    "Average Price": df['price'].mean(),
    "Most Common RoomType": df['room type'].mode(),
    #"Most Common Neighbourhood": df['neighbourhood'].mode()[0],
    "Highest Price": df['price'].max(),
    "Lowest Price": df['price'].min()
}
```

```python
print(summary)
```
